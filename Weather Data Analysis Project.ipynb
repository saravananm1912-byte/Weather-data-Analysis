```python
print("Shape of weather_df:", weather_df.shape)
print("\nInfo of weather_df:")
weather_df.info()
print("\nHead of weather_df:")
print(weather_df.head())
print("\nMissing values in weather_df:")
print(weather_df.isnull().sum())
```

```python
print("\nDescriptive statistics for numerical columns:")
print(weather_df.describe())

print("\nUnique values and counts for categorical columns:")
for column in weather_df.select_dtypes(include=['object', 'category']).columns:
    print(f"\nColumn: {column}")
    print("Unique Values:", weather_df[column].unique())
    print("Value Counts:\n", weather_df[column].value_counts())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

if weather_df.empty:
    print("weather_df is empty. Cannot generate a histogram for temperature distribution.")
else:
    # Assuming 'Temperature' column exists if weather_df was not empty
    if 'Temperature' in weather_df.columns:
        plt.figure(figsize=(10, 6))
        sns.histplot(weather_df['Temperature'], kde=True)
        plt.title('Distribution of Temperature Values')
        plt.xlabel('Temperature')
        plt.ylabel('Frequency')
        plt.grid(True)
        plt.show()
    else:
        print("The 'Temperature' column is not found in weather_df.")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

if weather_df.empty:
    print("weather_df is empty. Cannot generate a scatter plot for Temperature vs. Humidity.")
else:
    if 'Temperature' in weather_df.columns and 'Humidity' in weather_df.columns:
        plt.figure(figsize=(10, 6))
        sns.scatterplot(x=weather_df['Temperature'], y=weather_df['Humidity'])
        plt.title('Temperature vs. Humidity Relationship')
        plt.xlabel('Temperature')
        plt.ylabel('Humidity')
        plt.grid(True)
        plt.show()
    else:
        missing_cols = []
        if 'Temperature' not in weather_df.columns:
            missing_cols.append('Temperature')
        if 'Humidity' not in weather_df.columns:
            missing_cols.append('Humidity')
        print(f"The following required column(s) were not found in weather_df: {', '.join(missing_cols)}.")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

if weather_df.empty:
    print("weather_df is empty. Cannot generate a pie chart for seasonal weather distribution.")
else:
    if 'Season' in weather_df.columns:
        seasonal_counts = weather_df['Season'].value_counts()
        plt.figure(figsize=(8, 8))
        plt.pie(seasonal_counts, labels=seasonal_counts.index, autopct='%1.1f%%', startangle=140)
        plt.title('Seasonal Weather Distribution')
        plt.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.
        plt.legend(title="Seasons")
        plt.show()
    else:
        print("The 'Season' column is not found in weather_df. Please ensure seasons are derived from date columns first.")
```
